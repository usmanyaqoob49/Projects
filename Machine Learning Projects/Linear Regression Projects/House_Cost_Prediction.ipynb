```python
#Importing Libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
#Making a simple traing set

x_train = np.array([1.0,2.0])
y_train = np.array([300.0,500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
print(f"x_train shap {x_train.shape}")
m = len(x_train)
print(f"No of records {m}")
```

```python
#Displaying

i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
i = 1 
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}) , y^({i})) = ({x_i} , {y_i})")
```

```python
#Pplotting the Data set that will be used for training

plt.scatter(x_train,y_train,marker = 'x' , c = 'r')
plt.title("Housing Price")
plt.ylabel("price (in 1000s of dollars)")
plt.xlabel("size (1000 sqft)")
plt.show()
```

```python
#weights identified

w = 100
b = 100
print(f"w = {w}")
print(f"b = {b}")
```

```python
#Making Model
def compute_model_output(x, w, b):
    m = len(x)
    f_wb = np.zeros(m)
    for i in range (m):
        f_wb[i] = w * x[i] + b
    
    return f_wb
```

```python
#Passing training set to model
tmp_f_wb = compute_model_output(x_train, w, b)

#Plotting the model's Prediction
plt.plot(x_train, tmp_f_wb, c = 'b', label = 'Our Prediction')

#Plotting data points
plt.plot(x_train, y_train , marker = 'x', c = 'r', label = 'Actual Value')
plt.title("Housing Price")
plt.ylabel("price (in 1000s of dollars)")
plt.xlabel("size (1000 sqft)")
plt.legend()
plt.show()
```
